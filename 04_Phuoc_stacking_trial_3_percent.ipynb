```python
class RNN(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, num_classes):
        super(RNN, self).__init__()
        self.gru = nn.GRU(input_size = input_size, hidden_size = hidden_size, num_layers = 3, batch_first = True)
        self.foc = nn.Linear((hidden_size), (num_classes))

    def forward(self, x):
        x, h_n = self.gru(x)
        output = self.foc(h_n[-1])
        return output
```

```python
model = RNN(input_size, hidden_size, num_layers, num_classes).double().to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr = learning_rate)
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr = 0.0001)
```

```python
losses = []
print_loss_every = 20
print_acc_every = 20

for epoch in range(200):
    for i, (videos, labels, framenums) in enumerate(train_loader):
        information = torch.nn.utils.rnn.pack_padded_sequence(videos, framenums, batch_first = True, enforce_sorted = False).to(device)
        # origin shape: [4, 3, 32, 32] = 4, 3, 1024
        # input_layer: 3 input channels, 6 output channels, 5 kernel size
        labels = labels.to(torch.long).to(device)

        # Forward pass
        outputs = model(information)
        loss = criterion(outputs, labels)
        losses.append(loss)

        # Backward and optimize
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if (i+1) % print_loss_every == 0:
          print (f'Epoch [{epoch+1}/{num_epochs}], Step [{i+1}], Loss: {loss.item():.4f}')

    with torch.no_grad():
        total = 0
        total_correct = 0
        for i, (videos, labels, framenums) in enumerate(test_loader):
            information = torch.nn.utils.rnn.pack_padded_sequence(videos, framenums, batch_first = True, enforce_sorted = False).to(device)
            # origin shape: [4, 3, 32, 32] = 4, 3, 1024
            # input_layer: 3 input channels, 6 output channels, 5 kernel size
            labels = labels.to(torch.long).to(device)

            # Forward pass
            outputs = model(information)
            outputs = torch.argmax(outputs, dim=-1)

            # Backward and optimize
            total += len(outputs)
            total_correct += (outputs == labels).sum().item()

        print(f"Accuracy: {total_correct / total * 100:.2f}")
```
